```python
import re

def returnGeneratedAnswer(generatedAnswer, list_of_answer_choices):
    if (len(generatedAnswer) == 1):
        if (generatedAnswer.lower() == "a"):
            return "a"
        elif (generatedAnswer.lower() == "b"):
            return "b"
        elif (generatedAnswer.lower() == "c"):
            return "c"
        elif (generatedAnswer.lower() == "d"):
            return "d"
    else:
        if (generatedAnswer.lower().startswith('a.')):
            return "a"
        elif (generatedAnswer.lower().startswith('b.')):
            return "b"  
        elif (generatedAnswer.lower().startswith('c.')):
            return "c"
        elif (generatedAnswer.lower().startswith('d.')):
            return "d" 
        elif (generatedAnswer.lower() == "answer: a"):
            return "a"
        elif (generatedAnswer.lower() == "answer: b"):
            return "b"
        elif (generatedAnswer.lower() == "answer: c"):
            return "C"
        elif (generatedAnswer.lower() == "answer: d"):
            return "d"
        elif (generatedAnswer.lower() == "the answer is a"):
            return "a"
        elif (generatedAnswer.lower() == "the answer is b"):
            return "b"
        elif (generatedAnswer.lower() == "the answer is c"):
            return "c"
        elif (generatedAnswer.lower() == "the answer is d"):
            return "d"
        elif (generatedAnswer.lower() in list_of_answer_choices[0].lower()):
            return "a"
        elif (generatedAnswer.lower() in list_of_answer_choices[1].lower()):
            return "b"
        elif (generatedAnswer.lower() in list_of_answer_choices[2].lower()):
            return "c"
        elif (generatedAnswer.lower() in list_of_answer_choices[3].lower()):
            return "d"
        elif ( list_of_answer_choices[0].lower() in generatedAnswer.lower()):
            return "a"
        elif ( list_of_answer_choices[1].lower() in generatedAnswer.lower()):
            return "b"
        elif ( list_of_answer_choices[2].lower() in generatedAnswer.lower()):
            return "c"
        elif ( list_of_answer_choices[3].lower() in generatedAnswer.lower()):
            return "d"
        elif ("the correct answer is a" in generatedAnswer.lower()):
            return "a"
        elif ( "the correct answer is b" in generatedAnswer.lower()):
            return "b"
        elif ("the correct answer is c" in generatedAnswer.lower()):
            return "c"
        elif ( "the correct answer is d" in generatedAnswer.lower()):
            return "d"
        
    #all other cases
    return "?"

def checkEquivalencyOfTwoAnswer(correctAnswer, generatedAnswer, list_of_answer_choices):
    #return 1 if they are equivalent
    #return 0 if they are not equivalent
    #return 2 if unknown - to be determined manually
    
    adjustedGeneratedAnswer = returnGeneratedAnswer(generatedAnswer, list_of_answer_choices)
    if (adjustedGeneratedAnswer == "?"):
        #to be determined manually
        return 2 
    
    if (correctAnswer == returnGeneratedAnswer(generatedAnswer, list_of_answer_choices)):
        return 1
    else:
        return 0
```

```python
def extract_answer_choices(input_str):
    start_index = input_str.find("Answer Choices are:")
    if start_index == -1:
        return [] # return empty list if "Answer Choices are:" is not found in the input string
    start_index += len("Answer Choices are:")
    end_index = input_str.find("Specify their correct answer by choosing one of")
    if end_index == -1:
        return [] # return empty list if "Specify their correct answer by choosing one of" is not found in the input string
    answer_choices_str = input_str[start_index:end_index].strip() # extract the portion of the string between "Answer Choices are:" and "Specify their correct answer by choosing one of"
    answer_choices_list = answer_choices_str.split("\n") # split the answer choices string into a list of strings
    answer_choices_list = [choice.strip() for choice in answer_choices_list] # remove any leading/trailing whitespaces from each string in the list
    return answer_choices_list




#print(extract_answer_choices(df['question'][1000]))
```

```python
import pandas as pd
import json
import os

df = pd.read_csv("/content/drive/MyDrive/question-only MuMuQA/MuMuQA_questionOnly_all.csv", sep=",",  
                 usecols=['index', 'qid', 'question', 'origanswer', 'generatedanswer', 'generatedanswers'])

results = pd.DataFrame()
startindex = 0

numOfProblems = len(df['qid'])
print("the number of problems " + str(numOfProblems))


for ind in range(startindex,numOfProblems): 
    print("############")
    index1 = df['index'][ind]
    qid1 = df['qid'][ind]
    question1=df['question'][ind]
    answer1=df['origanswer'][ind]
    generatedanswer1=df['generatedanswer'][ind]
    generatedanswers1=df['generatedanswers'][ind]
    
    print("index: ", index1)
    print("qid: ", qid1)
    print("question: ", question1)
    print("answer: ", answer1)
    print("generatedanswer: ", generatedanswer1)
    #print("generatedanswers: ", generatedanswers1)
    
    list_of_answer_choices = extract_answer_choices(question1)
    correctness = checkEquivalencyOfTwoAnswer(answer1, generatedanswer1, list_of_answer_choices)
    print("correcness: ", correctness)

    dict2 = {'index': index1, 'qid': qid1, 'question': question1, 'origanswer': answer1, 'generatedanswer': generatedanswer1, 'correctness': correctness }

    results = results.append(dict2, ignore_index=True)
    #####in the loop

filename = "MuMuQA_questionOnly_withCorrectness_withAnswerCheck.csv"

results.to_csv(filename, index=False)

```
